```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pylab as plt
```

```python
diamonds = pd.read_csv('../data/diamonds_train.csv')
diamonds.head()
```

```python
diamonds.describe()
```

```python
cat_var_lst = ["city", "cut", "color", "clarity"]
num_var_lst = ([i for i in list(diamonds.columns) if i not in cat_var_lst])
```

```python
for col in num_var_lst:
    q1 = diamonds[col].quantile(0.25)
    q3 = diamonds[col].quantile(0.75)
    iqr = q3 - q1
    upper = q3 + 1.5 * iqr
    upper_array = np.array(diamonds[col] >= upper)
    lower = q1 - 1.5 * iqr
    lower_array = np.array(diamonds[col] <= lower)
    print(col, "Upper bound:", upper, "Lower bound:", lower)
    print(col, "Upper outliers", upper_array.sum(), "Lower outliers", lower_array.sum(), '\n')
```
